# üö¢ Titanic Dataset - Exploratory Data Analysis (EDA)

## üß≠ Overview
The Titanic dataset contains details of passengers aboard the RMS Titanic, which tragically sank in 1912.  
The goal of this analysis is to explore the data, identify patterns, and understand which factors influenced passenger survival.

---

## üîπ 1. Dataset Overview
- Total rows: **891**  
- Total columns: **12**  
- **Target variable:** `Survived` (0 = Did not survive, 1 = Survived)
- **Features include:**  
  - Demographics: `Age`, `Sex`, `Pclass`  
  - Travel info: `Embarked`, `Ticket`, `Fare`  
  - Family info: `SibSp`, `Parch`

---

## üîπ 2. Data Cleaning
- Missing values found in: `Age`, `Cabin`, and `Embarked`
  - `Age` ‚Üí Filled with **median**
  - `Embarked` ‚Üí Filled with **most frequent value (mode)**
  - `Cabin` ‚Üí Using feature engineering created a new column as has cabin
- Removed duplicate rows (if any)
- Checked and corrected data types

---

## üîπ 3. Univariate Analysis
- **Survival Count:** ~38% passengers survived  
- **Gender:** More males than females on board  
- **Pclass:** Most passengers were from **3rd class**  
- **Age:** Majority between **20‚Äì40 years**

---

## üîπ 4. Bivariate Analysis
- **Gender vs Survival:**  
  üßç‚Äç‚ôÄÔ∏è Females had a much higher survival rate than males  
- **Class vs Survival:**  
  üèÜ 1st class passengers had the highest survival rate  
- **Age vs Survival:**  
  üë∂ Younger passengers (especially children) were more likely to survive  
- **Embarked vs Survival:**  
  üö¢ Passengers from **Cherbourg (C)** had higher survival chances

---

## üîπ 7. Key Insights
‚úÖ Women and children had higher survival rates  
‚úÖ 1st class passengers were more likely to survive  
‚úÖ Higher fare = better chance of survival  
‚úÖ Most passengers embarked from **Southampton (S)**  

---

## üîπ 8. Conclusion
The Titanic dataset highlights that **social and economic factors** (like gender, class, and fare) played a major role in survival.  
This dataset is ideal for practicing **data visualization**, **feature engineering**, and **classification models** in data science.

#1.5) Handling missing Values

###a) Filling median value of age column in the missing values of the same column.

```python
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
```

###b) Filling mode value of Embarked column in place of missing value of same column.

```python
embarked_mode = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(embarked_mode)
```

##c) Creating a new column called "Has_cabin" where 0 indicates no cabin and 1 indicates has cabin.

```python
df['Has_Cabin'] = df['Cabin'].notna().astype(int)
```

```python
df.drop('Cabin', axis =1 , inplace = True)
```

##**a.) Does gender affect the chances of survival? (Survived vs Sex)**

##**b.) How does age influence survival rate? (Survived vs Agegroups)**

```python
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
(pd.crosstab(df['AgeGroup'], df['Survived'],  normalize = 'index')*100)
```

```python
(pd.crosstab(df['AgeGroup'], df['Survived'],  normalize = 'index')*100).plot(kind = 'bar', title = "Age group influence on survival count", figsize =(12,8) )
plt.show()
```
